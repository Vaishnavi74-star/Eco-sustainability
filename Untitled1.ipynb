```python
class User:
    _id_counter = 1

    def __init__(self, email, password):
        self.id = User._id_counter
        User._id_counter += 1
        self.email = email
        self.password = password
        self.products = []
        self.purchases = []
        self.cart = []

    def check_password(self, password):
        return self.password == password

    def __str__(self):
        return f"User(id={self.id}, email={self.email})"


class Product:
    _id_counter = 1

    def __init__(self, user_id, title, category, description, price):
        self.id = Product._id_counter
        Product._id_counter += 1
        self.user_id = user_id
        self.title = title
        self.category = category
        self.description = description
        self.price = price

    def __str__(self):
        return f"[{self.id}] {self.title} - ${self.price} ({self.category})"


class EcoFindsApp:
    def __init__(self):
        self.users = []
        self.products = []

    # ------------------ USER ------------------
    def register_user(self, email, password):
        if any(u.email == email for u in self.users):
            return "Email already exists!"
        user = User(email, password)
        self.users.append(user)
        return f"User registered: {user.email}"

    def login_user(self, email, password):
        for user in self.users:
            if user.email == email and user.check_password(password):
                return user
        return None

    # ------------------ PRODUCT ------------------
    def add_product(self, user, title, category, description, price):
        product = Product(user.id, title, category, description, price)
        self.products.append(product)
        user.products.append(product)
        return product

    def get_user_products(self, user):
        return user.products

    def get_all_products(self):
        return self.products

    def get_product_by_id(self, product_id):
        for p in self.products:
            if p.id == product_id:
                return p
        return None

    def search_products(self, keyword):
        return [p for p in self.products if keyword.lower() in p.title.lower()]

    def filter_by_category(self, category):
        return [p for p in self.products if p.category.lower() == category.lower()]

    # ------------------ CART ------------------
    def add_to_cart(self, user, product_id):
        product = self.get_product_by_id(product_id)
        if product:
            user.cart.append(product)
            return f"Added to cart: {product.title}"
        return "Product not found"

    def view_cart(self, user):
        return user.cart

    def checkout(self, user):
        user.purchases.extend(user.cart)
        user.cart = []
        return "Checkout complete."

    def view_purchases(self, user):
        return user.purchases


# ------------------ DEMO ------------------

if __name__ == "__main__":
    app = EcoFindsApp()

    # Register users
    print(app.register_user("john@example.com", "pass123"))
    print(app.register_user("jane@example.com", "secret"))

    # Login
    user = app.login_user("john@example.com", "pass123")
    if user:
        print(f"Logged in as {user.email}")
    else:
        print("Login failed.")

    # Add products
    app.add_product(user, "Eco Bottle", "Kitchen", "Reusable eco-friendly bottle", 10.99)
    app.add_product(user, "Organic T-Shirt", "Clothing", "Made from 100% organic cotton", 25.00)

    # View all products
    print("\nAll Products:")
    for p in app.get_all_products():
        print(p)

    # Search
    print("\nSearch 'T-Shirt':")
    for p in app.search_products("T-Shirt"):
        print(p)

    # Category filter
    print("\nFilter by 'Kitchen':")
    for p in app.filter_by_category("Kitchen"):
        print(p)

    # Cart operations
    app.add_to_cart(user, 1)
    print("\nCart:")
    for p in app.view_cart(user):
        print(p)

    # Checkout
    print("\n" + app.checkout(user))

    # Purchases
    print("\nPrevious Purchases:")
    for p in app.view_purchases(user):
        print(p)
```

```python
class User:
    _id_counter = 1

    def __init__(self, email, password):
        self.id = User._id_counter
        User._id_counter += 1
        self.email = email
        self.password = password
        self.products = []
        self.cart = []
        self.purchases = []

    def check_password(self, password):
        return self.password == password


class Product:
    _id_counter = 1

    def __init__(self, user_id, title, category, description, price):
        self.id = Product._id_counter
        Product._id_counter += 1
        self.user_id = user_id
        self.title = title
        self.category = category
        self.description = description
        self.price = price

    def __str__(self):
        return f"[{self.id}] {self.title} - ${self.price} ({self.category})"


class EcoFindsApp:
    def __init__(self):
        self.users = []
        self.products = []

    def register_user(self, email, password):
        if any(u.email == email for u in self.users):
            print("Email already registered!")
            return None
        user = User(email, password)
        self.users.append(user)
        print("User registered successfully.")
        return user

    def login_user(self, email, password):
        for user in self.users:
            if user.email == email and user.check_password(password):
                print("Login successful.")
                return user
        print("Invalid credentials.")
        return None

    def add_product(self, user, title, category, description, price):
        product = Product(user.id, title, category, description, price)
        self.products.append(product)
        user.products.append(product)
        print("Product added successfully.")

    def view_all_products(self):
        if not self.products:
            print("No products available.")
            return
        for p in self.products:
            print(p)

    def view_user_products(self, user):
        if not user.products:
            print("You haven't listed any products.")
            return
        for p in user.products:
            print(p)

    def search_products(self, keyword):
        results = [p for p in self.products if keyword.lower() in p.title.lower()]
        if not results:
            print("No matching products found.")
        for p in results:
            print(p)

    def filter_products(self, category):
        results = [p for p in self.products if p.category.lower() == category.lower()]
        if not results:
            print("No products in this category.")
        for p in results:
            print(p)

    def add_to_cart(self, user, product_id):
        product = next((p for p in self.products if p.id == product_id), None)
        if product:
            user.cart.append(product)
            print(f"Added '{product.title}' to cart.")
        else:
            print("Product not found.")

    def view_cart(self, user):
        if not user.cart:
            print("Your cart is empty.")
            return
        print("Your Cart:")
        for p in user.cart:
            print(p)

    def checkout(self, user):
        if not user.cart:
            print("Your cart is empty.")
            return
        user.purchases.extend(user.cart)
        user.cart.clear()
        print("Checkout successful.")

    def view_purchases(self, user):
        if not user.purchases:
            print("No previous purchases.")
            return
        print("Your Previous Purchases:")
        for p in user.purchases:
            print(p)


def main():
    app = EcoFindsApp()
    current_user = None

    while True:
        if current_user is None:
            print("\n--- EcoFinds ---")
            print("1. Register")
            print("2. Login")
            print("3. Exit")
            choice = input("Choose an option: ")

            if choice == '1':
                email = input("Email: ")
                password = input("Password: ")
                app.register_user(email, password)

            elif choice == '2':
                email = input("Email: ")
                password = input("Password: ")
                user = app.login_user(email, password)
                if user:
                    current_user = user

            elif choice == '3':
                print("Goodbye!")
                break

            else:
                print("Invalid option.")

        else:
            print(f"\nWelcome, {current_user.email}!")
            print("1. Add Product")
            print("2. View All Products")
            print("3. View My Listings")
            print("4. Search Products")
            print("5. Filter by Category")
            print("6. Add to Cart")
            print("7. View Cart")
            print("8. Checkout")
            print("9. View Previous Purchases")
            print("10. Logout")
            choice = input("Choose an option: ")

            if choice == '1':
                title = input("Product Title: ")
                category = input("Category: ")
                description = input("Description: ")
                price = float(input("Price: "))
                app.add_product(current_user, title, category, description, price)

            elif choice == '2':
                app.view_all_products()

            elif choice == '3':
                app.view_user_products(current_user)

            elif choice == '4':
                keyword = input("Enter keyword to search: ")
                app.search_products(keyword)

            elif choice == '5':
                category = input("Enter category to filter: ")
                app.filter_products(category)

            elif choice == '6':
                product_id = int(input("Enter Product ID to add to cart: "))
                app.add_to_cart(current_user, product_id)

            elif choice == '7':
                app.view_cart(current_user)

            elif choice == '8':
                app.checkout(current_user)

            elif choice == '9':
                app.view_purchases(current_user)

            elif choice == '10':
                current_user = None
                print("Logged out.")

            else:
                print("Invalid option.")


if __name__ == '__main__':
    main()

```
